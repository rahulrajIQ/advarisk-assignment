```python
# Renaming cols to be more clear
df = df.rename(columns={'ORDERNUMBER': 'order_nmber', 'QUANTITYORDERED':'quantity_ordered', 'PRICEEACH': 'price_per_item', 
                        'ORDERLINENUMBER':'order_line_no','ORDERDATE':'order_date',
       'PRODUCTLINE':'product_line','PRODUCTCODE':'product_code', 'CUSTOMERNAME': 'customer_name', 
       'ADDRESSLINE1':'address_line_1', 'ADDRESSLINE2':'address_line_2','POSTALCODE':'postal_code',
       'CONTACTLASTNAME':'contact_first_name', 'CONTACTFIRSTNAME':'contact_last_namme',
       'DEALSIZE':'deal_size'})
```

```python
# Changing all cols to lower case
df.columns = map(str.lower, df.columns)
```

```python
# order_date to datetime conversion
df.order_date = pd.to_datetime(df.order_date, format = 'mixed',dayfirst=True)
```

```python
# order_date dtype to str conversion
df['order_date'] =df['order_date'].astype(str)
```

```python
# concating address and dropping address_line_1 and address_line_2
df["address"] = df["address_line_1"].fillna(' ') + "  " + df["address_line_2"].fillna('')   
df.drop(['address_line_1','address_line_2'], axis=1, inplace=True)
```

```python
df.drop_duplicates(keep='first', inplace=True)
```

```python
import re
#Just removing . for now
list_of_chars = ['.']
# Create regex pattern to match all characters in list
pattern = '[' +  ''.join(list_of_chars) +  ']'
```

```python
df.phone= df.phone.apply(lambda x: re.sub(pattern,'', x))
```

```python
dt.datetime.today().strftime('%Y-%m-%d')
df.to_csv('sales_data_'+dt.datetime.today().strftime('%Y-%m-%d')+'.csv', index=False)
```

```python
def create_table_query(df):
    query = "CREATE TABLE IF NOT EXISTS " + 'sales_data' + ' ( ' + 'id serial PRIMARY KEY'

    for i in df.columns:
        query =  query + ', ' + i + ' varchar'

    query =  query + ' );'   
    
    return query

def insert_table_query(df):
    query = "INSERT INTO " + 'sales_data' + ' ( ' + (', ').join(df.columns) +') ' + 'VALUES ( ' 
    for i in df.columns:
        query =  query + '{}' + ',' 

    query =  query[:-1] + ') ' + 'ON CONFLICT DO NOTHING;'  
    return query
```

```python
import psycopg2
df = pd.read_csv('sales_data_'+dt.datetime.today().strftime('%Y-%m-%d')+'.csv')


def create_database():
    """
    - Creates and connects to the advarisk
    - Returns the connection and cursor to advarisk
    """
    
    # connect to default database
    conn = psycopg2.connect("host=127.0.0.1 dbname=postgres user=postgres password=xyz")
    conn.set_session(autocommit=True)
    cur = conn.cursor()
    
    # create advarisk database with UTF8 encoding
    cur.execute("DROP DATABASE IF EXISTS advarisk")
    cur.execute("CREATE DATABASE advarisk WITH ENCODING 'utf8' TEMPLATE template0")

    # close connection to default database
    conn.close()    
    
    # connect to advarisk database
    conn = psycopg2.connect("host=127.0.0.1 dbname=advarisk user=postgres password=xyz")
    cur = conn.cursor()
    
    return cur, conn



def main():
    """
    - Drops (if exists) and Creates the sparkify database. 
    
    - Establishes connection with the sparkify database and gets
    cursor to it.  
    
    - Drops all the tables.  
    
    - Creates all tables needed. 
    
    - Finally, closes the connection. 
    """
    cur, conn = create_database()
    
    cur.execute("DROP TABLE IF EXISTS songplays")
    conn.commit()
    
    
    cur.execute(create_table_query(df))
    conn.commit()
    
    insert_query = insert_table_query(df)
    for i in range(0,df.shape[0]):
        insert_query_updated= insert_query.format(df.iloc[i]['order_nmber'],
                                                 df.iloc[i]['quantity_ordered'],
                                                 df.iloc[i]['price_per_item'],
                                                 df.iloc[i]['order_line_no'],
                                                 df.iloc[i]['sales'],
                                                 df.iloc[i]['order_date'],
                                                 df.iloc[i]['status'],
                                                 df.iloc[i]['qtr_id'],
                                                 df.iloc[i]['month_id'],
                                                 df.iloc[i]['year_id'],
                                                 df.iloc[i]['product_line'],
                                                 df.iloc[i]['msrp'],
                                                 df.iloc[i]['product_code'],
                                                 df.iloc[i]['customer_name'],
                                                 df.iloc[i]['phone'],
                                                 df.iloc[i]['city'],
                                                 df.iloc[i]['state'],
                                                 df.iloc[i]['postal_code'],
                                                 df.iloc[i]['country'],
                                                 df.iloc[i]['territory'],
                                                 df.iloc[i]['contact_first_name'],
                                                 df.iloc[i]['contact_last_namme'],
                                                 df.iloc[i]['deal_size'],
                                                 df.iloc[i]['address'])
        cur.execute(insert_query_updated)
        conn.commit()
        print('Done row {}'.format(i))

    conn.close()


if __name__ == "__main__":
    main()
```
